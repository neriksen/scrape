# Dansk Initiativ for Etisk Handel: Brugen af kernebegreber inden for etisk handel på Twitter

## Kort projektbeskrivelse

I nærværende projekt undersøger vi udviklingen i brugen af begreber, der er relateret til etisk handel. Vi bruger Twitter-data til at undersøge udviklingen i den kvantitative anvendelse af begreberne for 2015 og frem. Eksempelvis ser vi nærmere på, om brugen af begrebet 'bæredygtig(hed)' på Twitter har været stigende eller faldende i perioden, idet udviklingen kan ses som en proxy for aktualiteten af politikområdet. 

```python
import GetOldTweets3 as got
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('bmh')
```

```python
tweet_criteria = got.manager.TweetCriteria().setQuerySearch('bæredygtig')\
                                            .setSince('2015-01-01')\

tweets = got.manager.TweetManager.getTweets(tweet_criteria)
```

```python
text_tweets = [[tweet.username,
              tweet.text, 
              tweet.date, 
              tweet.retweets,
              tweet.favorites,
              tweet.mentions,
              tweet.hashtags] for tweet in tweets]
    
bæredygtig_df = pd.DataFrame(text_tweets,
                            columns = ['User', 'Text', 'Date', 'Retweets', 'Favorites', 'Mentions', 'Hashtags'])

bæredygtig_df
```

```python
bæredygtig_df
```

```python
bæredygtig_df['Instances_bæredygtig'] = bæredygtig_df['Text'].str.count('bæredygtig')
```

```python
pd.to_datetime(bæredygtig_df['Date'], format='%d%b%Y:%H:%M:%S.%f')
```

```python
print(bæredygtig_df['Date'])
```

```python
gb_year = bæredygtig_df
gb_year.index = bæredygtig_df['Date']
gb_year = gb_year.groupby(by=[gb_year.index.year])

gb_year['Instances_bæredygtig'].sum().plot(kind='bar')
plt.show()
```
